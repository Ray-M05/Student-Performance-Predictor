# 1. Introducción

El rendimiento académico es un tema central en el ámbito educativo, ya que influye directamente en las oportunidades académicas y profesionales futuras de los estudiantes. Sin embargo, el desempeño en un examen no depende únicamente de la “inteligencia” o capacidad innata, sino de una combinación de factores académicos, familiares, personales y de estilo de vida.

En este proyecto se analiza un conjunto de datos de estudiantes que recoge información sobre hábitos de estudio, asistencia a clase, motivación, implicación de los padres, características del profesorado, entorno escolar y otros aspectos relacionados con su día a día. Junto con estas variables, se dispone además de la calificación final obtenida en un examen (`Exam_Score`), que utilizaremos como medida del rendimiento académico.

A partir de estos datos, se aplicarán distintas herramientas estadísticas para explorar qué factores parecen tener una mayor influencia sobre el rendimiento, con el objetivo de obtener una visión más clara de qué elementos podrían potenciar (o dificultar) el desempeño académico de los estudiantes.

## 1.1. Tema general

El tema general de este proyecto es el análisis de los factores que influyen en el rendimiento académico de los estudiantes, medido a través de la variable `Exam_Score`. Para ello se utilizará información sobre hábitos de estudio, contexto familiar, entorno escolar y estilo de vida.

En particular, se prestará especial atención a variables como las horas de estudio (`Hours_Studied`), la asistencia a clase (`Attendance`), el nivel de motivación (`Motivation_Level`), la calidad percibida del profesorado (`Teacher_Quality`), las horas de sueño (`Sleep_Hours`), la implicación de los padres (`Parental_Involvement`) y el tipo de escuela (`School_Type`). Nos interesa estudiar cómo se relacionan estos factores con la nota del examen y, en algunos casos, si generan diferencias importantes entre distintos grupos de estudiantes.

## 1.2. Objetivo general

El objetivo general del estudio es analizar estadísticamente el conjunto de datos *Student Performance Factors* para identificar y cuantificar la relación entre distintos factores académicos, familiares, personales y de entorno, y el rendimiento académico de los estudiantes medido por la variable `Exam_Score`.

## 1.3. Objetivos específicos

A partir de este objetivo general, se plantean varios objetivos específicos:

1. Describir la distribución de las principales variables del dataset (horas de estudio, asistencia, motivación, calidad del profesorado, implicación parental, tipo de escuela, etc.) y caracterizar el comportamiento de la variable de rendimiento `Exam_Score`.

2. Explorar, de forma gráfica y numérica, la relación entre variables académicas clave (como `Hours_Studied`, `Attendance` y `Previous_Scores`) y la calificación del examen.

3. Evaluar si existen diferencias estadísticamente significativas en el rendimiento (`Exam_Score`) entre grupos de estudiantes definidos por su nivel de motivación (`Motivation_Level`) y por la calidad percibida del profesorado (`Teacher_Quality`).

4. Construir un modelo de regresión lineal múltiple que permita explicar el rendimiento académico (`Exam_Score`) a partir de un conjunto seleccionado de variables explicativas (académicas, familiares, personales y del entorno escolar), e interpretar la dirección y magnitud de los efectos estimados.

```python
# ============================
# 2. Análisis exploratorio de datos (EDA)
# ============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# 2.1 Cargar datos y definir subconjunto de interés
df = pd.read_csv("StudentPerformanceFactors.csv")

cols_main = [
    "Exam_Score",
    "Hours_Studied",
    "Attendance",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Motivation_Level",
    "Teacher_Quality",
    "Sleep_Hours",
    "Parental_Involvement",
    "Family_Income",
    "School_Type",
    "Learning_Disabilities",
]

df_main = df[cols_main].copy()

print("Dimensiones de df_main:", df_main.shape)

# Clasificar variables
numeric_cols = [
    "Exam_Score",
    "Hours_Studied",
    "Attendance",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Sleep_Hours",
]

categorical_cols = [
    "Motivation_Level",
    "Teacher_Quality",
    "Parental_Involvement",
    "Family_Income",
    "School_Type",
    "Learning_Disabilities",
]

print("\nVariables numéricas:", numeric_cols)
print("Variables categóricas:", categorical_cols)

# Resumen de valores faltantes
print("\nValores faltantes por variable:")
print(df_main.isna().sum())
```

### 2.2 Análisis univariante

El análisis univariante se centra en estudiar cada variable por separado, con dos objetivos principales:

1. Resumir numéricamente la información (medias, medianas, dispersión, etc.).

A partir de este análisis será posible describir, por ejemplo, si las calificaciones de `Exam_Score` se concentran en un rango determinado, si la asistencia suele ser alta, o si las horas de estudio están equilibradas o muy desiguales entre estudiantes.


#### 2.2.1 Análisis univariante de variables numéricas

En esta sección se analizan las siguientes variables numéricas:

- `Exam_Score`
- `Hours_Studied`
- `Attendance`
- `Previous_Scores`
- `Tutoring_Sessions`
- `Sleep_Hours`

Para cada una de ellas se muestran:

- Estadísticos descriptivos básicos (media, desviación estándar, cuartiles, mínimo y máximo).
- Histogramas con estimación de densidad para visualizar la forma de la distribución.
- Diagramas de caja (*boxplots*) para identificar posibles valores atípicos.


```python
# ============================
# 2.2 Análisis univariante
# ============================

display(df_main[numeric_cols].describe())

def plot_numeric_univariate(data, columns):
    n = len(columns)
    fig, axes = plt.subplots(nrows=n, ncols=2, figsize=(12, 4 * n))
    
    if n == 1:
        axes = np.array([axes])
    
    for i, col in enumerate(columns):
        # Histograma + KDE
        sns.histplot(
            data=data,
            x=col,
            kde=True,
            ax=axes[i, 0]
        )
        axes[i, 0].set_title(f"Distribución de {col}")
        axes[i, 0].set_xlabel(col)
        
        # Boxplot
        sns.boxplot(
            data=data,
            x=col,
            ax=axes[i, 1]
        )
        axes[i, 1].set_title(f"Boxplot de {col}")
        axes[i, 1].set_xlabel(col)
    
    plt.tight_layout()
    plt.show()

plot_numeric_univariate(df_main, numeric_cols)
```

   - El histograma muestra una forma **aproximadamente uniforme**, aunque se observan algunos estudiantes que estudian muy poco (cercanos a 1–5 horas) y otros que estudian bastante más de lo habitual (por encima de 35–40 horas), marcados como **posibles outliers** en el boxplot.

Este análisis permite describir el perfil de la muestra en términos de:

- Niveles de motivación reportados (`Motivation_Level`).

```python
for col in categorical_cols:
    print(f"\nFrecuencias para {col}:")

    freq_abs = df_main[col].value_counts()
    freq_rel = df_main[col].value_counts(normalize=True) * 100
    freq_table = pd.concat([freq_abs, freq_rel], axis=1)
    freq_table.columns = ["Frecuencia", "Porcentaje (%)"]
    display(freq_table.round(2))
    
    plt.figure()
    sns.countplot(
        data=df_main,
        x=col,
        order=df_main[col].value_counts().index
    )
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

#### 2.3.1 Relación con `Hours_Studied` y `Attendance` (P1)

Para abordar la pregunta **P1**, se analizan las relaciones entre:

- `Exam_Score` y `Hours_Studied`.
- `Exam_Score` y `Attendance`.

Se utilizan:

- Gráficos de dispersión para observar visualmente si existe una tendencia creciente o decreciente.
- La correlación lineal entre `Exam_Score`, `Hours_Studied` y `Attendance` como medida numérica de la fuerza y dirección de la relación.


```python
# 2.3.1 Relación con Hours_Studied y Attendance (P1)

# Scatterplot Exam_Score vs Hours_Studied
plt.figure()
sns.scatterplot(
    data=df_main,
    x="Hours_Studied",
    y="Exam_Score",
    alpha=0.5
)
plt.title("Exam_Score vs Hours_Studied")
plt.xlabel("Hours_Studied")
plt.ylabel("Exam_Score")
plt.tight_layout()
plt.show()

# Scatterplot Exam_Score vs Attendance
plt.figure()
sns.scatterplot(
    data=df_main,
    x="Attendance",
    y="Exam_Score",
    alpha=0.5
)
plt.title("Exam_Score vs Attendance")
plt.xlabel("Attendance")
plt.ylabel("Exam_Score")
plt.tight_layout()
plt.show()

# Correlaciones simples
corr_p1 = df_main[["Exam_Score", "Hours_Studied", "Attendance"]].corr()
print("\nCorrelaciones simples (Exam_Score, Hours_Studied, Attendance):")
display(corr_p1)
```

### Observaciones: relación entre esfuerzo académico y rendimiento (P1)

1. **Relación entre `Exam_Score` y `Hours_Studied`**
   - En el diagrama de dispersión se aprecia una **tendencia claramente creciente**: a medida que aumentan las horas de estudio, las calificaciones tienden a ser más altas.
   - La correlación entre ambas variables es de aproximadamente **0.45**, lo que indica una **relación lineal positiva de intensidad moderada**.
   - Para valores muy bajos de horas de estudio (en torno a 5–10 horas), las notas se concentran mayoritariamente entre **60 y 70 puntos**, mientras que para niveles más altos (25–40 horas) es más frecuente observar calificaciones **por encima de 70**.
   - No obstante, existe una **dispersion considerable**: hay estudiantes que, con un número parecido de horas, obtienen resultados bastante distintos, lo que sugiere que el esfuerzo en horas de estudio **no es el único factor** que determina el rendimiento.

2. **Relación entre `Exam_Score` y `Attendance`**
   - El scatterplot muestra también una **tendencia creciente**: a mayores niveles de asistencia, las notas tienden a ser superiores.
   - La correlación entre `Exam_Score` y `Attendance` es de aproximadamente **0.58**, algo mayor que la de las horas de estudio, lo que sugiere una **relación positiva moderada-fuerte** entre asistencia y rendimiento.
   - Para asistencias cercanas al **60–70 %** se observan sobre todo notas en el rango **60–70**, mientras que cuando la asistencia se sitúa por encima del **90 %** aparecen con más frecuencia calificaciones **por encima de 70–75**.
   - De nuevo hay variabilidad dentro de cada nivel de asistencia, pero la **tendencia general** apunta a que una mayor presencia en clase se asocia con mejores notas.

3. **Relación entre `Hours_Studied` y `Attendance`**

- Low  → 1  
- Medium → 2  
- High → 3  

Las nuevas variables se guardan con el sufijo `_ord` para distinguirlas de las

```python
df_model = df_main.copy()
# Mapeo ordinal: Low=1, Medium=2, High=3
ordinal_map = {"Low": 1, "Medium": 2, "High": 3}

ordinal_cols = ["Motivation_Level", "Teacher_Quality",
                "Parental_Involvement", "Family_Income"]

for col in ordinal_cols:
    new_col = col + "_ord"
    df_model[new_col] = df_model[col].map(ordinal_map)

# Vista de ejemplo: variable original vs. codificada
df_model[["Motivation_Level", "Motivation_Level_ord"]].head()
```

## 4. Aplicación de técnicas estadísticas

En esta sección se aplican técnicas inferenciales para responder directamente
a las preguntas de investigación:

- **P2:** ¿Existen diferencias significativas en `Exam_Score` según
  `Motivation_Level` y `Teacher_Quality`? → ANOVA de un factor.
- **P1 y P3:** ¿Cómo se relacionan horas de estudio, asistencia y otros
  factores con `Exam_Score` cuando se consideran simultáneamente? → Regresión
  lineal múltiple.
- Como complemento, se aplica **Análisis de Componentes Principales (PCA)**
  sobre los predictores numéricos para explorar su estructura conjunta.

Se trabaja con los dataframes preparados en la etapa anterior:

- `df_main`: datos “limpios” con variables categóricas originales.
- `df_model`: datos listos para modelar.

```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt

# Comprobamos que df_main y df_model existen
print("Shape df_main:", df_main.shape)
print("Shape df_model:", df_model.shape)
```

### 4.1 Pruebas de hipótesis para P2 (ANOVA)

Para responder a **P2** se estudia si la media de `Exam_Score` cambia según:

- los niveles de `Motivation_Level` (Low / Medium / High), y
- los niveles de `Teacher_Quality` (Low / Medium / High).

En cada caso se aplica un **ANOVA de un factor** con nivel de significación
$\alpha = 0.05$.

#### 4.1.1 Hipótesis

Para cada variable categórica (por ejemplo, `Motivation_Level`):

- $H_0$: las medias de `Exam_Score` son iguales en todos los grupos  
  (por ejemplo, $\mu_{\text{Low}} = \mu_{\text{Medium}} = \mu_{\text{High}}$).
- $H_1$: al menos un grupo presenta una media diferente.

Se interpretan el estadístico F y el p-valor, y se comparan las medias por
grupo para describir las diferencias observadas.


```python
# ANOVA de un factor: Exam_Score ~ Motivation_Level
model_mot = smf.ols("Exam_Score ~ C(Motivation_Level)", data=df_main).fit()
anova_mot = sm.stats.anova_lm(model_mot, typ=2)

print("ANOVA: Exam_Score ~ Motivation_Level")
display(anova_mot)

print("\nMedias de Exam_Score por nivel de Motivation_Level:")
df_main.groupby("Motivation_Level")["Exam_Score"].agg(["mean", "std", "count"])
```

Los resultados del ANOVA para `Motivation_Level` muestran:

- Un estadístico F considerablemente grande (≈ 25) y
- un p-valor muy pequeño (p < 0.001),

por lo que se **rechaza $H_0$**: hay evidencia estadísticamente significativa
de que la media de `Exam_Score` **no es la misma** para todos los niveles de
motivación.

Al observar las medias por grupo se ve que:

- Los estudiantes con **High** motivación presentan, en promedio,
  una nota de examen ligeramente superior a los de **Medium** y **Low**.
- Las diferencias en términos de puntos son relativamente pequeñas
  (del orden de ≈ 0.5–1 punto), pero estadísticamente significativas debido
  al gran tamaño muestral.

Conclusión parcial: mayor motivación se asocia con un rendimiento algo mejor,
aunque el efecto en puntos no es gigantesco.


```python
# ANOVA de un factor: Exam_Score ~ Teacher_Quality
model_tq = smf.ols("Exam_Score ~ C(Teacher_Quality)", data=df_main).fit()
anova_tq = sm.stats.anova_lm(model_tq, typ=2)

print("ANOVA: Exam_Score ~ Teacher_Quality")
display(anova_tq)

print("\nMedias de Exam_Score por nivel de Teacher_Quality:")
df_main.groupby("Teacher_Quality")["Exam_Score"].agg(["mean", "std", "count"])
```

En el ANOVA para `Teacher_Quality` se observa también:

- Un estadístico F notable (≈ 20) y
- un p-valor muy pequeño (p < 0.001),

por lo que se **rechaza $H_0$** y se concluye que la media de `Exam_Score`
**depende del nivel de calidad percibida del profesorado**.

Las medias por grupo indican que:

- El grupo con `Teacher_Quality = High` presenta, en promedio, una nota algo
  superior al de `Medium` y al de `Low`.
- De nuevo, las diferencias son de menos de 1 punto entre extremos, pero
  consistentes y estadísticamente significativas.

Conclusión parcial: tanto la motivación del estudiante como la calidad del
profesorado se relacionan con diferencias (moderadas) en el rendimiento
académico, respondiendo afirmativamente a P2 en el plano estadístico.


### 4.2 Regresión lineal múltiple para P1 y P3

Para responder conjuntamente a **P1** y **P3** se construye un modelo de
**regresión lineal múltiple** con:

- Variable dependiente: `Exam_Score`.
- Predictores (desde `df_model`):

  - Variables de esfuerzo y desempeño previo:
    `Hours_Studied`, `Attendance`, `Previous_Scores`,
    `Tutoring_Sessions`, `Sleep_Hours` (todas estandarizadas).
  - Variables ordinales codificadas:
    `Motivation_Level_ord`, `Teacher_Quality_ord`,
    `Parental_Involvement_ord`, `Family_Income_ord`.
  - Variables nominales codificadas:
    `School_Type_Private`, `Learning_Disabilities_Yes`.

En notación simplificada, el modelo es:

$$
Exam\_Score_i = \beta_0
 + \beta_1 \,\text{Hours\_Studied}_i
 + \beta_2 \,\text{Attendance}_i
 + \dots
 + \beta_{11} \,\text{Learning\_Disabilities\_Yes}_i
 + \varepsilon_i
$$

Con este modelo se busca:

- Cuantificar el efecto de `Hours_Studied` y `Attendance` sobre `Exam_Score`
  (**P1**) controlando por el resto de factores.
- Identificar qué variables son los predictores más importantes del
  rendimiento (**P3**).


```python
# Fórmula del modelo de regresión lineal múltiple
formula = """
Exam_Score ~ Hours_Studied + Attendance + Previous_Scores
            + Tutoring_Sessions + Sleep_Hours
            + Motivation_Level_ord + Teacher_Quality_ord
            + Parental_Involvement_ord + Family_Income_ord
            + School_Type_Private + Learning_Disabilities_Yes
"""

model = smf.ols(formula, data=df_model).fit()
model.summary()
```

```python
# Métricas globales del modelo
metrics = pd.Series({
    "R2": model.rsquared,
    "R2_ajustado": model.rsquared_adj,
    "F_stat": model.fvalue,
    "p_valor_F": model.f_pvalue
})
print("Métricas globales del modelo:\n")
display(metrics)

# Tabla de coeficientes con p-valores
coef_table = pd.DataFrame({
    "coef": model.params,
    "std_err": model.bse,
    "t": model.tvalues,
    "p_value": model.pvalues
})
coef_table
```

#### 4.2.1 Calidad global del modelo

- El coeficiente de determinación $R^2$ es aproximadamente **0.66**,
  lo que indica que el modelo explica en torno al **66% de la variabilidad**
  observada en `Exam_Score`.
- El estadístico F es muy grande y el p-valor global del modelo
  (p ≈ 0) es prácticamente nulo, por lo que se rechaza la hipótesis
  de que “todos los coeficientes (salvo la intercepción) sean cero”.

Conclusión: el modelo de regresión lineal múltiple ofrece **un ajuste
razonablemente bueno** para explicar el rendimiento académico.

#### 4.2.2 Efecto de `Hours_Studied` y `Attendance` (P1)

Los coeficientes estimados para:

- `Hours_Studied` (estandarizada) son positivos y altamente significativos
  (p < 0.001).  
  → Un aumento de 1 desviación estándar en horas de estudio se asocia, en
  promedio, con un **incremento de alrededor de 1.7–1.8 puntos** en el
  examen, manteniendo constantes los demás factores.

- `Attendance` (estandarizada) también presenta coeficiente positivo y muy
  significativo (p < 0.001).  
  → Un aumento de 1 desviación estándar en asistencia se asocia con un
  incremento de alrededor de **2.3 puntos** en `Exam_Score`.

Esto responde a **P1**: tanto estudiar más como asistir más a clase se
relaciona fuertemente con un mejor rendimiento, incluso tras controlar por
otras variables.

#### 4.2.3 Predictores relevantes adicionales (P3)

Otros coeficientes significativos (p < 0.05) incluyen:

- `Previous_Scores` y `Tutoring_Sessions`: ambos con efecto positivo,
  indicando que el desempeño previo y el uso de tutorías se relacionan con

  - `Motivation_Level_ord`, `Teacher_Quality_ord` y `Family_Income_ord`
    también influyen positivamente, aunque con magnitud algo menor.

### 4.3 Análisis de Componentes Principales (PCA) 

Como técnica complementaria, se aplica **PCA** sobre un subconjunto de
predictores numéricos, con el objetivo de:

- Reducir la dimensionalidad del conjunto de predictores.
- Identificar componentes que agrupen variables en posibles “factores”
  interpretables (por ejemplo, esfuerzo académico, contexto familiar, etc.).

Se utilizan las siguientes variables (todas numéricas en `df_model`):

- `Hours_Studied`, `Attendance`, `Previous_Scores`,
  `Tutoring_Sessions`, `Sleep_Hours`,
- `Motivation_Level_ord`, `Teacher_Quality_ord`,
  `Parental_Involvement_ord`, `Family_Income_ord`,
- `School_Type_Private`, `Learning_Disabilities_Yes`.

Antes del PCA se estandarizan de nuevo estas variables para que todas tengan
media 0 y varianza 1.


```python
pca_features = [
    "Hours_Studied", "Attendance", "Previous_Scores",
    "Tutoring_Sessions", "Sleep_Hours",
    "Motivation_Level_ord", "Teacher_Quality_ord",
    "Parental_Involvement_ord", "Family_Income_ord",
    "School_Type_Private", "Learning_Disabilities_Yes"
]

X_pca = df_model[pca_features].values

# Estandarización para PCA
scaler_pca = StandardScaler()
X_scaled = scaler_pca.fit_transform(X_pca)

# PCA sin fijar número de componentes (usamos todos)
pca = PCA()
X_pca_transformed = pca.fit_transform(X_scaled)

explained_var = pd.DataFrame({
    "Componente": [f"PC{i+1}" for i in range(len(pca_features))],
    "Varianza_explicada": pca.explained_variance_ratio_,
    "Varianza_acumulada": pca.explained_variance_ratio_.cumsum()
})

explained_var
```

```python
plt.figure(figsize=(6, 4))
plt.plot(
    range(1, len(pca_features) + 1),
    pca.explained_variance_ratio_,
    marker="o"
)
plt.xlabel("Componente principal")
plt.ylabel("Proporción de varianza explicada")
plt.title("Gráfico de sedimentación (scree plot)")
plt.xticks(range(1, len(pca_features) + 1))
plt.show()
```

```python
loadings = pd.DataFrame(
    pca.components_.T,
    index=pca_features,
    columns=[f"PC{i+1}" for i in range(len(pca_features))]
)

# Mostramos las cargas de los tres primeros componentes
loadings.iloc[:, :3].round(3)
```

El PCA muestra que:

- Cada componente principal explica alrededor de un **9–10%** de la varianza.
- La varianza acumulada crece de forma bastante uniforme:
  - Las primeras 3 componentes explican en torno a un 28–30%,
  - Las primeras 5 alrededor de un 47%,
  - Las primeras 7 cerca de un 65%.

Es decir, **no existe un único componente dominante** que concentre la mayor
parte de la variabilidad: la información está repartida de forma relativamente
homogénea entre los predictores.

Al observar las **cargas**:

- Algunas componentes agrupan sobre todo variables académicas
  (`Previous_Scores`, `Hours_Studied`, `Attendance`, `Tutoring_Sessions`),
  lo que podría interpretarse como un factor de “esfuerzo y desempeño
  académico”.
- Otras combinan variables de contexto familiar y personal
  (`Family_Income_ord`, `Parental_Involvement_ord`,
  `Learning_Disabilities_Yes`, etc.), más relacionadas con la situación del
  estudiante.

El PCA no se usa aquí como modelo principal, sino como **herramienta
exploratoria complementaria** para entender mejor la estructura de los
predictores utilizados en la regresión.

### 5.2 Conclusión para P1: horas de estudio, asistencia y rendimiento

**P1.** ¿Cómo se relacionan las horas de estudio y la asistencia con
  el rendimiento académico (`Exam_Score`)?

En relación con **P1**, los resultados del modelo de regresión indican que tanto
las horas de estudio como la asistencia a clase están claramente asociadas con

(como las notas previas, la motivación, la implicación de los padres, el nivel
de ingresos o la calidad del profesorado), los coeficientes de
`Hours_Studied` y `Attendance` siguen siendo positivos y altamente
significativos. 

En términos prácticos, los estudiantes que estudian más horas y que asisten
con mayor regularidad tienden a obtener notas más altas en el examen. La
asistencia aparece como un predictor ligeramente más fuerte que las horas de
estudio, lo que sugiere que no solo importa estudiar horas por cuenta propia,
sino también estar presente en clase, seguir las explicaciones y mantener un
contacto continuo con el ritmo del curso. En conjunto, los resultados apoyan
claramente la idea de que **estudiar más y faltar menos** se relaciona con un
mejor rendimiento.


### 5.3 Conclusión para P2: motivación, calidad del profesorado y diferencias en la nota

**P2.** ¿Existen diferencias significativas en `Exam_Score` según el

Respecto a **P2**, los ANOVA aplicados muestran que sí existen diferencias

ambos casos se rechaza la hipótesis de igualdad de medias, lo que indica que
la calificación en `Exam_Score` depende, al menos en parte, de estas
variables.

Al comparar las medias por grupo se observa una tendencia clara: los
estudiantes con **alta motivación** obtienen, en promedio, una nota algo
superior a la de aquellos con motivación media o baja; de forma similar,
quienes perciben a su profesorado como de **alta calidad** presentan un
rendimiento medio algo mayor que los alumnos de centros percibidos como de
calidad media o baja. Las diferencias en puntos no son enormes, pero son
consistentes y, dado el gran tamaño muestral, suficientemente claras como para
afirmar que tanto la motivación como la calidad docente **marcan una
diferencia real** en el rendimiento académico.


### 5.4 Conclusión para P3: factores que mejor explican el rendimiento

**P3.** ¿Qué factores explican mejor el rendimiento académico cuando se
  consideran simultáneamente en un modelo de regresión múltiple?

En el caso de **P3**, el modelo de regresión múltiple permite ver el panorama
“con todo a la vez”, es decir, qué factores explican mejor el rendimiento
académico cuando se consideran simultáneamente. El hecho de que el modelo
explique alrededor de dos tercios de la variabilidad en `Exam_Score` ya indica
que las variables incluidas capturan buena parte de la historia.

Entre los predictores más relevantes destacan la asistencia (`Attendance`),
las horas de estudio (`Hours_Studied`), las notas previas (`Previous_Scores`)
y las sesiones de tutoría (`Tutoring_Sessions`): los estudiantes que ya venían
con buen desempeño, que estudian más, que acuden a clase y que aprovechan
recursos de apoyo tienden a rendir mejor. A esto se suma el bloque de
variables familiares y motivacionales: mayor implicación de los padres
(`Parental_Involvement_ord`), mejores condiciones económicas
(`Family_Income_ord`), mayor motivación (`Motivation_Level_ord`) y mejor
valoración del profesorado (`Teacher_Quality_ord`) se asocian también con

Por el lado negativo, declarar dificultades de aprendizaje
(`Learning_Disabilities_Yes`) está relacionado con un rendimiento algo menor,
incluso controlando por el resto de factores. En conjunto, la conclusión para
P3 es que el rendimiento académico no depende de un solo factor aislado,
sino de la **combinación** entre hábitos de estudio, contexto familiar, apoyo
académico y características del entorno escolar.


### 5.6 Conclusión general

En conjunto, el análisis sugiere que el rendimiento académico, medido a
través de `Exam_Score`, está fuertemente relacionado con los **hábitos de
estudio** (horas de estudio y asistencia), el **historial académico**
(notas previas y uso de tutorías) y diversos factores del **contexto
familiar y motivacional** (implicación de los padres, ingresos, motivación,
percepción de la calidad del profesorado y presencia o ausencia de
dificultades de aprendizaje).

Los resultados de los ANOVA confirman que existen diferencias significativas
en la nota media según el nivel de motivación y la calidad del profesorado,
mientras que el modelo de regresión múltiple muestra que la combinación de
estos factores, junto con otros de esfuerzo y contexto, explica una parte muy
importante de la variabilidad en el rendimiento. El PCA refuerza la idea de
que el fenómeno es multidimensional y que no hay un único “factor mágico”
que lo explique todo.

En resumen, mejorar el rendimiento académico no pasa únicamente por “estudiar
más”, sino por una estrategia integral que combine buenos hábitos de estudio
y asistencia, apoyo familiar, acceso a recursos académicos (como tutorías) y
un entorno educativo de calidad.

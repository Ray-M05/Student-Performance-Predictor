# 1. Introducción

El rendimiento académico es un tema central en el ámbito educativo, ya que influye directamente en las oportunidades académicas y profesionales futuras de los estudiantes. Sin embargo, el desempeño en un examen no depende únicamente de la “inteligencia” o capacidad innata, sino de una combinación de factores académicos, familiares, personales y de estilo de vida.

En este proyecto se analiza un conjunto de datos de estudiantes que recoge información sobre hábitos de estudio, asistencia a clase, motivación, implicación de los padres, características del profesorado, entorno escolar y otros aspectos relacionados con su día a día. Junto con estas variables, se dispone además de la calificación final obtenida en un examen (`Exam_Score`), que utilizaremos como medida del rendimiento académico.

A partir de estos datos, se aplicarán distintas herramientas estadísticas para explorar qué factores parecen tener una mayor influencia sobre el rendimiento, con el objetivo de obtener una visión más clara de qué elementos podrían potenciar (o dificultar) el desempeño académico de los estudiantes.

## 1.1. Tema general

El tema general de este proyecto es el análisis de los factores que influyen en el rendimiento académico de los estudiantes, medido a través de la variable `Exam_Score`. Para ello se utilizará información sobre hábitos de estudio, contexto familiar, entorno escolar y estilo de vida.

En particular, se prestará especial atención a variables como las horas de estudio (`Hours_Studied`), la asistencia a clase (`Attendance`), el nivel de motivación (`Motivation_Level`), la calidad percibida del profesorado (`Teacher_Quality`), las horas de sueño (`Sleep_Hours`), la implicación de los padres (`Parental_Involvement`) y el tipo de escuela (`School_Type`). Nos interesa estudiar cómo se relacionan estos factores con la nota del examen y, en algunos casos, si generan diferencias importantes entre distintos grupos de estudiantes.

## 1.2. Objetivo general

El objetivo general del estudio es analizar estadísticamente el conjunto de datos *Student Performance Factors* para identificar y cuantificar la relación entre distintos factores académicos, familiares, personales y de entorno, y el rendimiento académico de los estudiantes medido por la variable `Exam_Score`.

Evaluar si existen diferencias estadísticamente significativas en el rendimiento (`Exam_Score`) entre grupos de estudiantes definidos por su nivel de motivación (`Motivation_Level`) y por la calidad percibida del profesorado (`Teacher_Quality`). Construir un modelo de regresión lineal múltiple que permita explicar el rendimiento académico (`Exam_Score`) a partir de un conjunto seleccionado de variables explicativas (académicas, familiares, personales y del entorno escolar), e interpretar la dirección y magnitud de los efectos estimados.

### 2.1 Carga de datos y definición del subconjunto de interés

En primer lugar se carga el dataset original `StudentPerformanceFactors.csv` y se selecciona un subconjunto de variables que resultan especialmente relevantes para el estudio del rendimiento académico:

- `Exam_Score`: variable respuesta (calificación del examen).
- Variables académicas: `Hours_Studied`, `Attendance`, `Previous_Scores`, `Tutoring_Sessions`.
- Variables personales y de estilo de vida: `Sleep_Hours`, `Motivation_Level`.
- Variables del entorno escolar: `Teacher_Quality`, `School_Type`.
- Variables familiares y socioeconómicas: `Parental_Involvement`, `Family_Income`.

Además, se clasifica cada variable como numérica o categórica y se revisan los valores faltantes.  
Esta primera vista general permite detectar posibles problemas de calidad de datos (por ejemplo, categorías raras o `NaN` en `Teacher_Quality`) y confirmar que el subconjunto elegido es adecuado para responder a las preguntas de investigación P1–P3.


```python
# ============================
# 2. Análisis exploratorio de datos (EDA)
# ============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# 2.1 Cargar datos y definir subconjunto de interés
df = pd.read_csv("StudentPerformanceFactors.csv")

cols_main = [
    "Exam_Score",
    "Hours_Studied",
    "Attendance",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Motivation_Level",
    "Teacher_Quality",
    "Sleep_Hours",
    "Parental_Involvement",
    "Family_Income",
    "School_Type",
    "Learning_Disabilities",
]

df_main = df[cols_main].copy()

print("Dimensiones de df_main:", df_main.shape)

# Clasificar variables
numeric_cols = [
    "Exam_Score",
    "Hours_Studied",
    "Attendance",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Sleep_Hours",
]

categorical_cols = [
    "Motivation_Level",
    "Teacher_Quality",
    "Parental_Involvement",
    "Family_Income",
    "School_Type",
    "Learning_Disabilities",
]

print("\nVariables numéricas:", numeric_cols)
print("Variables categóricas:", categorical_cols)

# Resumen de valores faltantes
print("\nValores faltantes por variable:")
print(df_main.isna().sum())
```

### 2.2 Análisis univariante

El análisis univariante se centra en estudiar cada variable por separado, con dos objetivos principales:

1. Resumir numéricamente la información (medias, medianas, dispersión, etc.).

A partir de este análisis será posible describir, por ejemplo, si las calificaciones de `Exam_Score` se concentran en un rango determinado, si la asistencia suele ser alta, o si las horas de estudio están equilibradas o muy desiguales entre estudiantes.


#### 2.2.1 Análisis univariante de variables numéricas

En esta sección se analizan las siguientes variables numéricas:

- `Exam_Score`
- `Hours_Studied`
- `Attendance`
- `Previous_Scores`
- `Tutoring_Sessions`
- `Sleep_Hours`

Para cada una de ellas se muestran:

- Estadísticos descriptivos básicos (media, desviación estándar, cuartiles, mínimo y máximo).
- Histogramas con estimación de densidad para visualizar la forma de la distribución.
- Diagramas de caja (*boxplots*) para identificar posibles valores atípicos.

Este análisis permite responder preguntas como:

- ¿La distribución de `Exam_Score` es aproximadamente simétrica o está sesgada hacia calificaciones bajas/altas?
- ¿Las horas de estudio (`Hours_Studied`) se concentran en pocos valores o hay mucha variabilidad entre estudiantes?
- ¿Existen indicios de estudiantes con asistencia muy baja o muy alta respecto al resto (`Attendance`)?


```python
# 2.2 Análisis univariante
# ------------------------

# 2.2.1 Variables numéricas: tabla de resumen
display(df_main[numeric_cols].describe())

def plot_numeric_univariate(data, columns):
    n = len(columns)
    fig, axes = plt.subplots(nrows=n, ncols=2, figsize=(12, 4 * n))
    
    # Por si hubiera solo una variable
    if n == 1:
        axes = np.array([axes])
    
    for i, col in enumerate(columns):
        # Histograma + KDE
        sns.histplot(
            data=data,
            x=col,
            kde=True,
            ax=axes[i, 0]
        )
        axes[i, 0].set_title(f"Distribución de {col}")
        axes[i, 0].set_xlabel(col)
        
        # Boxplot
        sns.boxplot(
            data=data,
            x=col,
            ax=axes[i, 1]
        )
        axes[i, 1].set_title(f"Boxplot de {col}")
        axes[i, 1].set_xlabel(col)
    
    plt.tight_layout()
    plt.show()

plot_numeric_univariate(df_main, numeric_cols)
```

### Observaciones

1. **Exam_Score**

   - El histograma tiene forma aproximadamente **normal**, con algunos valores extremos altos (por encima de 90–95 e incluso mayores a 100), que se reflejan como **outliers** en el boxplot.

2. **Hours_Studied**
   - Los estudiantes estudian en promedio unas **20 horas**, con mediana también en **20**, lo que indica de nuevo cierta **simetría** en la distribución.
   - El 50 % central de los valores se sitúa entre **16 y 24 horas**, por lo que la mayoría estudia entre **unas 2 y 3 horas diarias** si pensamos en una semana.
   - El histograma muestra una forma **aproximadamente campaniforme**, aunque se observan algunos estudiantes que estudian muy poco (cercanos a 1–5 horas) y otros que estudian bastante más de lo habitual (por encima de 35–40 horas), marcados como **posibles outliers** en el boxplot.

3. **Attendance**
   - La asistencia media es cercana al **80 %**, con mediana también en **80 %**, y un rango intercuartílico de **70 % a 90 %**.
   - La distribución es relativamente **uniforme** dentro del rango 60–100 %, sin una única moda muy marcada, lo que indica que hay bastante diversidad en los niveles de asistencia.
   - El boxplot muestra una dispersión moderada y no se aprecian outliers muy extremos, lo que sugiere que la mayoría de estudiantes mantiene una asistencia razonablemente alta.

4. **Previous_Scores**
   - La media de calificaciones previas es de **75 puntos**, con mediana también **75**, y una desviación estándar relativamente alta (**≈14.4**), lo que indica **bastante variabilidad** entre estudiantes.
   - El 50 % central de los valores se concentra entre **63 y 88 puntos**, es decir, desde un rendimiento previo medio-bajo hasta uno claramente alto.
   - El histograma parece más **plano/extendido** que el de `Exam_Score`, con valores repartidos entre 50 y 100, lo que sugiere que los antecedentes académicos son más heterogéneos que la nota del examen actual.

5. **Tutoring_Sessions**
   - El número medio de sesiones de tutoría es de aproximadamente **1.5 sesiones**, con mediana **1**, lo que indica que **la mayoría de estudiantes acude a 0–2 sesiones**.
   - La distribución está claramente **sesgada a la derecha**: hay muchos estudiantes con pocas sesiones y pocos alumnos con muchas (valores de 5 a 8), que aparecen como **outliers** en el boxplot.

   - El histograma es relativamente **simétrico** alrededor de las 7 horas, aunque se observan algunos casos con **pocas horas de sueño (4–5)** y otros con **muchas (9–10)**, que podrían considerarse extremos pero no excesivamente alejados del resto de la distribución.

En conjunto, las variables numéricas presentan en general distribuciones bastante razonables, con:
- **`Exam_Score` y `Hours_Studied`** cercanas a una forma normal y con concentración fuerte alrededor de sus medias.
- **`Attendance` y `Previous_Scores`** más dispersas, mostrando mayor heterogeneidad entre estudiantes.
- **`Tutoring_Sessions`** claramente sesgada, indicando que la mayoría usa poco este recurso.

#### 2.3.1 Relación con `Hours_Studied` y `Attendance` (P1)

Para abordar la pregunta **P1**, se analizan las relaciones entre:

- `Exam_Score` y `Hours_Studied`.
- `Exam_Score` y `Attendance`.

Se utilizan:

- Gráficos de dispersión para observar visualmente si existe una tendencia creciente o decreciente.
- La correlación lineal entre `Exam_Score`, `Hours_Studied` y `Attendance` como medida numérica de la fuerza y dirección de la relación.

Aspectos a comentar (una vez vistos los gráficos y la matriz de correlación):

- Si se observa que, en promedio, más horas de estudio se asocian con mayores calificaciones.
- Si la asistencia a clase parece estar fuertemente vinculada al rendimiento o si la relación es más débil.
- Posibles patrones no lineales o presencia de grupos diferenciados de estudiantes.


```python
# 2.3.1 Relación con Hours_Studied y Attendance (P1)

# Scatterplot Exam_Score vs Hours_Studied
plt.figure()
sns.scatterplot(
    data=df_main,
    x="Hours_Studied",
    y="Exam_Score",
    alpha=0.5
)
plt.title("Exam_Score vs Hours_Studied")
plt.xlabel("Hours_Studied")
plt.ylabel("Exam_Score")
plt.tight_layout()
plt.show()

# Scatterplot Exam_Score vs Attendance
plt.figure()
sns.scatterplot(
    data=df_main,
    x="Attendance",
    y="Exam_Score",
    alpha=0.5
)
plt.title("Exam_Score vs Attendance")
plt.xlabel("Attendance")
plt.ylabel("Exam_Score")
plt.tight_layout()
plt.show()

# Correlaciones simples
corr_p1 = df_main[["Exam_Score", "Hours_Studied", "Attendance"]].corr()
print("\nCorrelaciones simples (Exam_Score, Hours_Studied, Attendance):")
display(corr_p1)
```

### Observaciones: relación entre esfuerzo académico y rendimiento (P1)

1. **Relación entre `Exam_Score` y `Hours_Studied`**
   - En el diagrama de dispersión se aprecia una **tendencia claramente creciente**: a medida que aumentan las horas de estudio, las calificaciones tienden a ser más altas.
   - La correlación entre ambas variables es de aproximadamente **0.45**, lo que indica una **relación lineal positiva de intensidad moderada**.
   - Para valores muy bajos de horas de estudio (en torno a 5–10 horas), las notas se concentran mayoritariamente entre **60 y 70 puntos**, mientras que para niveles más altos (25–40 horas) es más frecuente observar calificaciones **por encima de 70**.
   - No obstante, existe una **dispersion considerable**: hay estudiantes que, con un número parecido de horas, obtienen resultados bastante distintos, lo que sugiere que el esfuerzo en horas de estudio **no es el único factor** que determina el rendimiento.

2. **Relación entre `Exam_Score` y `Attendance`**
   - El scatterplot muestra también una **tendencia creciente**: a mayores niveles de asistencia, las notas tienden a ser superiores.
   - La correlación entre `Exam_Score` y `Attendance` es de aproximadamente **0.58**, algo mayor que la de las horas de estudio, lo que sugiere una **relación positiva moderada-fuerte** entre asistencia y rendimiento.
   - Para asistencias cercanas al **60–70 %** se observan sobre todo notas en el rango **60–70**, mientras que cuando la asistencia se sitúa por encima del **90 %** aparecen con más frecuencia calificaciones **por encima de 70–75**.
   - De nuevo hay variabilidad dentro de cada nivel de asistencia, pero la **tendencia general** apunta a que una mayor presencia en clase se asocia con mejores notas.

3. **Relación entre `Hours_Studied` y `Attendance`**
   - La correlación entre `Hours_Studied` y `Attendance` es prácticamente **nula** (≈ -0.01), lo que indica que ambas variables son **casi independientes** en la muestra.
   - Es decir, no parece que los estudiantes que asisten más a clase estudien sistemáticamente más (o menos) horas por su cuenta; son dos dimensiones del esfuerzo académico **relativamente separadas**.

En conjunto, los resultados apoyan la idea de que tanto las **horas de estudio** como la **asistencia** influyen positivamente en el rendimiento, siendo la asistencia la que presenta una relación algo más fuerte con la calificación del examen.


#### 2.3.2 Relación con `Motivation_Level` y `Teacher_Quality` (P2)

Para explorar la pregunta **P2**, se estudian las diferencias en `Exam_Score` entre grupos definidos por:

- `Motivation_Level`
- `Teacher_Quality`

Se representan boxplots de `Exam_Score` para cada categoría, lo que permite:

- Comparar visualmente las medianas de rendimiento entre niveles de motivación.
- Evaluar si los estudiantes que perciben una mayor calidad del profesorado obtienen, en general, mejores resultados.
- Observar la dispersión dentro de cada grupo y la posible presencia de valores extremos.

Más adelante, estas diferencias se contrastarán formalmente mediante pruebas de hipótesis (por ejemplo, ANOVA) para determinar si las variaciones observadas son estadísticamente significativas.


```python
# 2.3.2 Relación con Motivation_Level y Teacher_Quality (P2)

plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df_main,
    x="Motivation_Level",
    y="Exam_Score",
    order=sorted(df_main["Motivation_Level"].dropna().unique())
)
plt.title("Exam_Score por nivel de Motivation_Level")
plt.xlabel("Motivation_Level")
plt.ylabel("Exam_Score")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df_main,
    x="Teacher_Quality",
    y="Exam_Score",
    order=sorted(df_main["Teacher_Quality"].dropna().unique())
)
plt.title("Exam_Score por nivel de Teacher_Quality")
plt.xlabel("Teacher_Quality")
plt.ylabel("Exam_Score")
plt.tight_layout()
plt.show()
```

### Observaciones: diferencias por motivación y calidad del profesorado (P2)

   - Los rangos intercuartílicos (la “altura” de las cajas) son similares entre niveles, lo que indica una **variabilidad comparable** en el rendimiento dentro de cada grupo de motivación.
   - En los tres niveles aparecen múltiples **outliers por encima de 80–90 puntos**, lo que muestra que incluso estudiantes con motivación baja pueden obtener calificaciones altas y, al revés, también hay estudiantes muy motivados con notas más moderadas.
   - En conjunto, la motivación parece asociarse **solo de forma moderada** con el rendimiento: se intuye una mejora ligera al pasar de motivación baja a media/alta, pero con **mucho solapamiento** entre grupos.

2. **Exam_Score según Teacher_Quality**

   - Al comparar los niveles de `Teacher_Quality` (`Low`, `Medium`, `High`), las **medianas** de `Exam_Score` también se encuentran en torno a los **67–68 puntos**.
   - Los grupos con **Teacher_Quality = High** y **Medium** muestran medianas y cuartiles superiores **ligeramente mayores** que el grupo **Low**, lo que sugiere que una mejor percepción de la calidad docente podría asociarse con **mejores resultados medios**.
   - El grupo de calidad **Low** presenta una distribución algo más “contenida” en la parte alta (tercer cuartil algo más bajo), mientras que en **High** y **Medium** se observan más valores por encima de **75–80 puntos**, reflejados en los outliers.
   - De nuevo, las distribuciones están **muy solapadas**: hay estudiantes con buenos resultados incluso cuando la calidad percibida del profesorado es baja, y viceversa.

3. **Conclusión preliminar para P2**

   - Tanto el nivel de **motivación** como la **calidad del profesorado** parecen influir en el rendimiento, pero las diferencias observadas en los boxplots son **relativamente sutiles** y con gran solapamiento entre grupos.
   - Estos resultados sugieren un posible efecto positivo de la motivación y de una mayor calidad docente sobre `Exam_Score`, pero será necesario aplicar **pruebas de hipótesis (por ejemplo, ANOVA)** para determinar si las diferencias entre grupos son **estadísticamente significativas** o podrían atribuirse al azar.

Finalmente, se calcula y representa la **matriz de correlaciones** entre las principales variables numéricas:

- `Exam_Score`

- `Attendance`

- `Tutoring_Sessions`

La visualización (por ejemplo, mediante un mapa de calor) permite:

- Identificar qué variables están más fuertemente correlacionadas con `Exam_Score`.
- Detectar posibles relaciones fuertes entre predictores (colinealidad), lo que será relevante para la construcción del modelo de regresión múltiple (P3).
- Sugerir combinaciones de variables que podrían resumir “esfuerzo académico”, “hábitos de estudio” o “bienestar”, que se retomarán en el análisis de componentes principales (PCA).

Esta sección cierra el EDA proporcionando una visión integrada de cómo se relacionan las variables numéricas entre sí y con el rendimiento académico.


```python
# 2.3.3 Matriz de correlaciones entre variables numéricas

corr_matrix = df_main[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1
)
plt.title("Matriz de correlaciones (variables numéricas)")
plt.tight_layout()
plt.show()
```

### Observaciones: matriz de correlaciones (2.3.3)

1. **Factores más relacionados con `Exam_Score`**
   - Las correlaciones más altas con `Exam_Score` son:
     - `Attendance`: **0.58**
     - `Hours_Studied`: **0.45**
   - Ambos coeficientes indican una **relación lineal positiva moderada**, algo más fuerte en el caso de la asistencia. Esto refuerza lo observado en los diagramas de dispersión: **asistir más a clase y estudiar más horas suele asociarse con mejores calificaciones**.
   - El resto de variables numéricas muestran correlaciones bastante menores con la nota:
     - `Previous_Scores`: ≈ **0.18**
     - `Tutoring_Sessions`: ≈ **0.16**
     - `Sleep_Hours`: ≈ **0** (−0.02)
   - Por tanto, los antecedentes académicos y las sesiones de tutoría parecen tener solo una **relación débil** con el rendimiento en este examen, mientras que las horas de sueño prácticamente **no muestran relación lineal** con la calificación.

2. **Relaciones entre predictores**
   - La correlación entre `Hours_Studied` y `Attendance` es prácticamente **nula** (≈ −0.01), lo que indica que ambas dimensiones del esfuerzo académico son **casi independientes**: hay estudiantes que asisten mucho pero estudian poco por su cuenta, y viceversa.
   - Las correlaciones entre el resto de pares de predictores (`Previous_Scores`, `Tutoring_Sessions`, `Sleep_Hours`) también son muy bajas (en torno a 0, ±0.02).
   - Esta falta de correlaciones fuertes entre predictores sugiere que, de cara a un modelo de **regresión lineal múltiple (P3)**, no se espera un problema grave de **multicolinealidad**: cada variable aporta información relativamente distinta.

3. **Implicaciones para el modelado posterior**
   - A la vista de la matriz, `Attendance` y `Hours_Studied` se perfilan como **candidatas principales** a incluir en el modelo explicativo de `Exam_Score`.
   - `Previous_Scores` y `Tutoring_Sessions` podrían aportar algo de información adicional, aunque su efecto esperado será más **modesto**.
   - `Sleep_Hours`, al no mostrar correlación lineal con la nota, podría tener un papel menos relevante en un modelo lineal simple, aunque no se descarta que existan **relaciones no lineales o efectos de interacción** que se podrían explorar más adelante.

En resumen, la matriz de correlaciones confirma que el **esfuerzo académico actual (asistencia y horas de estudio)** es el conjunto de factores numéricos que mantiene una relación más clara con el rendimiento en el examen.
